```python
#BeautifulSoup 불러오기
import requests
from bs4 import BeautifulSoup
```

```python
#크롤링할 웹사이트 설정
url = 'https://www.korea.kr/news/top50List.do'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)

top_news = []

#BeautifulSoup이 어떻게 되어있는지 보고 Parsing 적용할 수 있어야 함 
#크롤링할 태그 조건 설정
strong = soup.select("span.text > strong") # text라는 클래스명을 가진 span 태그의 자식 중 strong 태그만 선택
for n in strong:
    top_news.append(n.get_text().strip())

# Version(Date)가 중요 -> 처리 방법: 해당 version or date 출력하기 

from datetime import datetime
year = datetime.today().year
month = datetime.today().month
day = datetime.today().day

print(f'<<< {year}년 {month}월 {day}일', end=' ')
print('대한민국 정책 브리핑 TOP 50 뉴스 >>>')

n = 1
for i in top_news[:50]:
    print("%d: "%n + i)
    n = n+1

#크롤링 결과 확인
print(top_news)
```

```python
file_name = "먼 후일.txt"
fp = open(file_name, "r", encoding = "utf-8") #읽기용

#파일 읽기
data = fp.readlines() #리스트로 저장
fp.close() 

print(data)
```

```python
for line in data:
    print(line, end='')
```

```python
#복사본 파일 생성하고 열기
fp2 = open(file_name.split(".")[0] + "_작업본.txt", "w")

cnt = 0
where = []

for i, line in enumerate(data):
    if "잊었노라" in line:
        cnt += 1
        where.append(i)
        fp2.write(str(i) + " : " + line) #한 줄씩 쓰기

fp.close() 

print(f"'잊었노라'는 {where}줄에 총 {cnt}번 존재합니다.")
```
